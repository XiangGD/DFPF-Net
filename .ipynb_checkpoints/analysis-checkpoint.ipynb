```python
import os
import numpy as np
from PIL import Image
```

```python
def casename2idx(casename):
    cls = ['c', 's', 'i']
    if casename[0] in cls:    
        idx = cls.index(casename[0])
    return int(idx)
```

```python
def read_samples(data_dir,list_dir,analysis):
    list_path = os.path.join(list_dir,analysis+'_score_detection.txt')
    save_path = os.path.join(list_dir,analysis+'_samples')
    os.makedirs(save_path,exist_ok=True)
    os.makedirs(os.path.join(save_path,'image'),exist_ok=True)
    os.makedirs(os.path.join(save_path,'label'),exist_ok=True)
    sample_list = open(list_path).readlines()
    num_sample = [0,0,0]
    n1 = [0,0,0]
    message = []
    for sample_name in sample_list:
        indx=casename2idx(sample_name)
        num_sample[indx] += 1 
        
        data_path = os.path.join(data_dir, sample_name.strip('\n')+'.npz')
        data = np.load(data_path)
        image, label = data['image'], data['label']
        h,w = label.shape
        
        fake_area = np.sum(label)
        fake_scale = fake_area/(h*w+1e-6)
        message.append(sample_name.strip('\n')+'==>fake_area:'+str(fake_area)+'  fake_scale:'+str(fake_scale)+'\r\n')
        if fake_area <4000:
            n1[indx] += 1 
            print(sample_name,w,h,w*h,fake_area, (fake_area/(w*h)))
        image = Image.fromarray(np.uint8(image))
        label = Image.fromarray(np.uint8(label*255))
        image.save(os.path.join(save_path,'image', sample_name.strip('\n')+'.tif'))
        label.save(os.path.join(save_path,'label', sample_name.strip('\n')+'.tif'))
    with open(os.path.join(save_path, analysis+'_score_analyse.txt'), 'w') as f:
        f.writelines(message)          
    return num_sample, n1
```

```python
list_dir='./test_log/FU_Nist16256'
data_dir='../datasets/Nist16/Nist16_npz1'

zero_num_sample, n1 = read_samples(data_dir,list_dir,'zero')
print('c_num,s_num,re_num:',zero_num_sample[0],zero_num_sample[1],zero_num_sample[2],n1)
```

```python
low_num_sample,n2 = read_samples(data_dir,list_dir,'low')
print('c_num,s_num,re_num:',low_num_sample[0],low_num_sample[1],low_num_sample[2],n2)
```

```python
rig_num_sample, n3= read_samples(data_dir,list_dir,'right')
print('c_num,s_num,re_num:',rig_num_sample[0],rig_num_sample[1],rig_num_sample[2],n3)
```
