```python
import os
import cv2 as cv
import numpy as np 
from PIL import Image
```

```python
def label2binary(filename):
    #print('mask_path:',filename)
    mask = cv.imread(filename)
    mask = cv.cvtColor(mask, cv.COLOR_BGR2GRAY)
    ret,binary = cv.threshold(mask,127,255,cv.THRESH_BINARY)
    return binary
```

```python
def get_list(root_path):
    with open(root_path, "r") as f:  # 打开文件
        lines = f.readlines()  # 读取文件
    print('1',len(lines))   
    return lines
```

```python
# resize scale:0.78,0.25
def _resize(img_root,label_root, save_path, scale):
    img_path = os.path.join(save_path,'image_resize'+str(scale))
    label_path = os.path.join(save_path,'label_resize'+str(scale))
    os.makedirs(img_path,exist_ok=True )
    os.makedirs(label_path,exist_ok=True )
    img_name = img_root.split('/')[-1]
    label_name= label_root.split('/')[-1]
    img = Image.open(img_root)   
    label = Image.open(label_root)
    # Importing Image class from PIL module 

    width, height = img.size 
   
    h = np.int(height*scale)
    w = np.int(width*scale)
    resize_img = img.resize((w,h))
    resize_label= label.resize((w,h))
    #print(img.size,type(img),img_name)
 
    resize_img.save(os.path.join(img_path,img_name))
    resize_label.save(os.path.join(label_path,label_name))
```

```python
#noise siigma:3,5,15
def _gauss_noise(img_root, save_path, sigma=3, mean=0):
    img_path = os.path.join(save_path,'noise'+str(sigma))
    os.makedirs(img_path,exist_ok=True )
    img_name = img_root.split('/')[-1]
    img = cv.imread(img_root)
    
    noiseGause = np.random.normal(mean, sigma, img.shape)
    imgGaussNoise = img + noiseGause
    imgGaussNoise = np.clip(imgGaussNoise,a_min=0, a_max=255)
    #imgGaussNoise = np.uint8(cv.normalize(imgGaussNoise, None, 0, 255, cv.NORM_MINMAX))
    cv.imwrite(os.path.join(img_path,img_name),imgGaussNoise)
    
```

```python
#blur size:3,5,15
def _gauss_blur(img_root, save_path, size, mean=0): 
    img_path = os.path.join(save_path,'blur'+str(size))
    os.makedirs(img_path,exist_ok=True )
    img_name = img_root.split('/')[-1]
    img = cv.imread(img_root)
    
    blur_img = cv.GaussianBlur(img, (size, size), mean)#(5, 5)表示高斯矩阵的长与宽都是5，标准差取0
    cv.imwrite(os.path.join(img_path,img_name),blur_img)
```

```python
#jpeg quality:50,100
def _jpg_compress(img_root, save_path, quality):
    img_path = os.path.join(save_path,'compress'+str(quality))
    os.makedirs(img_path,exist_ok=True )
    img_name = img_root.split('/')[-1].replace('tif','jpg')
    img = cv.imread(img_root)

    cv.imwrite(os.path.join(img_path,img_name),img, [cv.IMWRITE_JPEG_QUALITY, quality])
    
```
